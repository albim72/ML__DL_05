# Tworzenie zbioru danych -> Zbi√≥r HIGGSA

```python
gz = tf.keras.utils.get_file('HIGGS.csv.gz','http://mlphysics.ics.uci.edu/data/higgs/HIGGS.csv.gz')
```

```python
FEATURES = 28
```

```python
ds = tf.data.experimental.CsvDataset(gz,[float(),]*(FEATURES+1),compression_type="GZIP")
```

```python
def pack_row(*row):
  label = row[0]
  features = tf.stack(row[1:],1)
  return features,label
```

```python
packed_ds = ds.batch(10000).map(pack_row).unbatch()
```

```python
for features,label in packed_ds.batch(1000).take(1):
  print(features[0])
  plt.hist(features.numpy().flatten(),bins=101)
```

```python
N_VALIDATION = int(1e3)
N_TRAIN = int(1e4)
BUFFER_SIZE = int(1e4)
BATCH_SIZE = 500
STEPS_PER_EPOCH = N_TRAIN/BATCH_SIZE
```

```python
validate_ds = packed_ds.take(N_VALIDATION).cache()
train_ds = packed_ds.skip(N_VALIDATION).take(N_TRAIN).cache()
```

```python
train_ds
```

```python
validate_ds = validate_ds.batch(BATCH_SIZE)
train_ds = train_ds.shuffle(BUFFER_SIZE).repeat().batch(BATCH_SIZE)
```

# PRZYGOTOWANIE MODELU DO PREZENTACJI PRZETRENOWANIA

```python
lr_schedule = tf.keras.optimizers.schedules.InverseTimeDecay(
    0.001,
    decay_steps = STEPS_PER_EPOCH*1000,
    decay_rate = 1,
    staircase = False
)

def get_optimizer():
  return tf.keras.optimizers.Adam(lr_schedule)
```

```python
step = np.linspace(0,100_000)
lr = lr_schedule(step)
plt.figure(figsize=(8,6))
plt.plot(step/STEPS_PER_EPOCH,lr)
plt.ylim([0,max(plt.ylim())])
plt.xlabel('Epoch')
_ = plt.ylabel('Learning Rate')
```

```python
def get_callbacks(name):
  return [
      tfdocs.modeling.EpochDots(),
      tf.keras.callbacks.EarlyStopping(monitor='val_binary_crossentropy',patience=200),
      tf.keras.callbacks.TensorBoard(logdir/name),
  ]
```

```python
def compile_and_fit(model,name,optimizer=None,max_epochs=10000):
  if optimizer is None:
    optimizer = get_optimizer()
  model.compile(optimizer=optimizer,
                loss = tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=[
                    tf.keras.losses.BinaryCrossentropy(
                        from_logits=True, name='binary_crossentropy'
                    ),'accuracy'
                ])
  model.summary()
  history = model.fit(
      train_ds,
      steps_per_epoch = STEPS_PER_EPOCH,
      epochs=max_epochs,
      validation_data=validate_ds,
      callbacks = get_callbacks(name),
      verbose=0
  )
  return history
```

# MODEL1 - BARDZO MA≈ÅY MODEL -> TINY

```python
from keras.src.metrics.confusion_metrics import activations
tiny_model = tf.keras.Sequential([
    layers.Dense(16,activation='elu',input_shape=(FEATURES,)),
    layers.Dense(1)
])
```

```python
size_histories = {}
```

```python
size_histories['Tiny'] = compile_and_fit(tiny_model,'sizes/Tiny')
```

```python
print(size_histories)
```

```python
plotter = tfdocs.plots.HistoryPlotter(metric='binary_crossentropy',smoothing_std=10)
plotter.plot(size_histories)
```

# MODEL2  MA≈ÅY MODEL -> SMALL

```python
small_model = tf.keras.Sequential([
    layers.Dense(16,activation='elu',input_shape=(FEATURES,)),
    layers.Dense(16,activation='elu'),
    layers.Dense(1)
])
```

```python
size_histories['Small'] = compile_and_fit(small_model,'sizes/Small')
```

# MODEL3 - ≈öREDNI MODEL -> MEDIUM

```python
medium_model = tf.keras.Sequential([
    layers.Dense(64,activation='elu',input_shape=(FEATURES,)),
    layers.Dense(64,activation='elu'),
    layers.Dense(64,activation='elu'),
    layers.Dense(1)
])
```

```python
size_histories['Medium'] = compile_and_fit(medium_model,'sizes/Medium')
```

# MODEL4 - DU≈ªY MODEL -> LARGE

```python
large_model = tf.keras.Sequential([
    layers.Dense(512,activation='elu',input_shape=(FEATURES,)),
    layers.Dense(512,activation='elu'),
    layers.Dense(512,activation='elu'),
    layers.Dense(512,activation='elu'),
    layers.Dense(1)
])
```

```python
size_histories['Large'] = compile_and_fit(large_model,'sizes/Large')
```

```python
plotter.plot(size_histories)
a = plt.xscale('log')
plt.xlim([5,max(plt.xlim())])
plt.ylim([0.5,0.7])
plt.xlabel("Epochs(skala logarytmiczna)")
```

# METODY PRZECIWDZIA≈ÅANIA PRZETRENOWANIU MODELU

```python
shutil.rmtree(logdir/'regularizers/Tiny',ignore_errors=True)
shutil.copytree(logdir/'sizes/Tiny', logdir/'regularizers/Tiny')
```

```python
regularizers_histories = {}
regularizers_histories['Tiny'] = size_histories['Tiny']
```

```python
#regularuzacja wag: L1,L2 - L2  mutacja waga, kt√≥rej koszt jest proporcjonalny do kwadratu wsp√≥≈Çczynnik√≥w wag
l2_model = tf.keras.Sequential([
    layers.Dense(512,activation='elu',kernel_regularizer=regularizers.l2(0.001),input_shape=(FEATURES,)),
    layers.Dropout(0.3),
    layers.Dense(512,activation='elu',kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.3),
    layers.Dense(512,activation='elu',kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.3),
    layers.Dense(512,activation='elu',kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.3),
    layers.Dense(1)
])
regularizers_histories['l2'] = compile_and_fit(l2_model,'regularizers/l2')
```

```python
plotter.plot(regularizers_histories)
plt.ylim([0.5,0.7])
```
